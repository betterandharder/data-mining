### Province Totals

```python
fig, ax = plt.subplots(figsize=(14, 6))

reviews_df.groupby(reviews_df['Country'])['Price'].mean().plot.bar()
mean_price = reviews_df["Price"].mean(0)
reviews_df["Price"].fillna(mean_price)

# Creates labels
ax.set_xlabel('Country')
ax.set_ylabel('Mean Price')
# Sets the plot title
ax.set_title("Mean Price per Country")
ax.legend(bbox_to_anchor=(1, 1), loc=2, title='Price')

fig.tight_layout()
```

# 3.0 Conclusion

### 结论

- 美国的俄勒冈州和加利福尼亚州是世界闻名的葡萄酒生产州。
- 并不是越贵的葡萄酒评分就越高，在10.00 - 20.00 价格区间的葡萄酒往往更能够获得更好的评分。
- 葡萄酒生产商确实是更应该注重客户的需求，而不是一味的追求更高的质量，因为更高的质量意味着更高的成本和更高的出售价格，反而不会受到欢迎。
- 在大多数商业葡萄酒时，价格和质量之间其实没有相关性。
